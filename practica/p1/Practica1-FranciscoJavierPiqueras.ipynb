```python
import datetime
import re
import spacy
import spacy.cli
```

```python
! wget https://www.clips.uantwerpen.be/conll2002/ner/data/esp.testb -P ./src
```

```python
raw_text = open('./src/esp.testb', 'r')
raw_text
```

```python
full_text = ''
word_category = {}
```

```python
for i, row in enumerate(raw_text):
    cat_element = re.split('\s', row)
    word_category[i] = (cat_element[0], cat_element[1])
    full_text += cat_element[0] + ' '
raw_text.close()
```

```python
spacy.cli.download("es")
model = spacy.load('es')
```

```python
tokens = model(full_text)
```

```python
print(tokens[1].text)
print(tokens[1].ent_iob_)
print(tokens[1].ent_type_)
```

```python
token_output = {}
for i, token in enumerate(tokens):
    token_output[i] = (token.text, token.ent_iob_ + "-" + token.ent_type_ 
                       if token.ent_iob_ != 'O' 
                       else token.ent_iob_)
```

```python
for i in token_output:
    if(token_output[i][0]!=word_category[i][0]):
        print(i, token_output[i][0], word_category[i][0])
```

#### Generamos el fichero de salida

```python
output_f = open("./src/output", "w")
```

```python
len(word_category)
```

```python
len(token_output)
```

```python
j = 0

for i in word_category:
    word_ = word_category[i][0]
    cat_ = word_category[i][1]
    
    if(word_.strip() != token_output[j][0].strip()):
        j+=1;
        print('The tokens ' + word_.strip() + ' and ' + token_output[j][0].strip() + ' do not match.')
    else:
        res_ = '' if word_.strip() == '' else token_output[i][1]
        output_f.write('{} {} {}\n'.format(word_, cat_, res_))
        
    j+=1;
    
output_f.close()
```
